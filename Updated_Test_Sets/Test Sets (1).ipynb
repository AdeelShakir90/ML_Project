```python
# Function for dataset exploration
def explore_dataset(df, name):
    print(f"\n Exploring {name} Dataset\n")
    
    # Display first few rows
    print("First 5 rows:")
    display(df.head())

    # Check for missing values
    print("\nMissing Values:")
    print(df.isnull().sum())

    # Get dataset statistics
    print("\nDataset Statistics:")
    display(df.describe())

    # Identify categorical and numerical columns
    categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
    numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()

    print("\nCategorical Columns:", categorical_cols)
    print("Numerical Columns:", numerical_cols)
    print("="*50)

# Explore each dataset (Assuming they are already loaded)
explore_dataset(ccd_train, "Credit Card Default (CCD)")
explore_dataset(kdd_train, "KDD")
explore_dataset(pid_train, "PIMA")
```

```python
from sklearn.preprocessing import LabelEncoder, MinMaxScaler

# Function to preprocess a dataset
def preprocess_dataset(df, name):
    print(f"\nProcessing {name} Dataset...\n")

    # Encode categorical variables
    label_encoders = {}
    categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
    
    for col in categorical_cols:
        le = LabelEncoder()
        df[col] = le.fit_transform(df[col])
        label_encoders[col] = le  # Save encoder for future use

    # Normalize numerical variables
    numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
    scaler = MinMaxScaler()
    df[numerical_cols] = scaler.fit_transform(df[numerical_cols])

    print(f"{name} Dataset Processed Successfully!")
    return df, label_encoders, scaler

# Apply preprocessing
ccd_train, ccd_encoders, ccd_scaler = preprocess_dataset(ccd_train, "Credit Card Default (CCD)")
kdd_train, kdd_encoders, kdd_scaler = preprocess_dataset(kdd_train, "KDD")
pima_train, pima_encoders, pima_scaler = preprocess_dataset(pid_train, "PIMA")

```

```python
import pandas as pd
import joblib
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Load the transformed train datasets
ccd_train = pd.read_csv(r"E:\Study Data MS Data Scince\ML(Machine Learning)\ML\CCD_train_LT.csv")
kdd_train = pd.read_csv(r"E:\Study Data MS Data Scince\ML(Machine Learning)\ML\KDD_train_LT.csv")
pima_train = pd.read_csv(r"E:\Study Data MS Data Scince\ML(Machine Learning)\ML\PID_train_LT.csv")

# Function to create and save encoders and scalers
def save_encoders_scalers(train_df, name):
    print(f"\nProcessing {name} Training Dataset...")

    # Encode categorical variables
    encoders = {}
    categorical_cols = train_df.select_dtypes(include=["object"]).columns
    for col in categorical_cols:
        le = LabelEncoder()
        train_df[col] = le.fit_transform(train_df[col])
        encoders[col] = le

    # Save encoders
    joblib.dump(encoders, f"{name.lower()}_encoders.pkl")

    # Normalize numerical variables
    scaler = StandardScaler()
    numerical_cols = train_df.select_dtypes(include=["int64", "float64"]).columns.tolist()
    train_df[numerical_cols] = scaler.fit_transform(train_df[numerical_cols])

    # Save scaler
    joblib.dump(scaler, f"{name.lower()}_scaler.pkl")

    print(f"{name} Training Dataset Encoders & Scalers Saved Successfully!")

# Apply encoding and scaling on training datasets
save_encoders_scalers(ccd_train, "CCD")
save_encoders_scalers(kdd_train, "KDD")
save_encoders_scalers(pima_train, "PIMA")

print("\nAll Encoders & Scalers have been saved successfully!")
```
